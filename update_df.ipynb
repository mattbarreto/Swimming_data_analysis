```python
# Crear un dataframe temporal a partir de los datos existentes
```

```python
# Convertir datos a formato compatible con el dataframe original
```

```python
temp[['Hora_inicio', 'Hora_de_finalización', 'Duración_sesion', 'Tiempo_estilo_espalda', 'Tiempo_estilo_pecho', 'Tiempo_estilo_crawl', 'Descanso_entre_series', 'Tiempo_ingesta_previa', 'Tiempo_ejercicios_previos']] = temp[['Hora_inicio', 'Hora_de_finalización', 'Duración_sesion', 'Tiempo_estilo_espalda', 'Tiempo_estilo_pecho', 'Tiempo_estilo_crawl', 'Descanso_entre_series', 'Tiempo_ingesta_previa', 'Tiempo_ejercicios_previos']].apply(pd.to_datetime, format='%H:%M:%S', errors='coerce')
```
